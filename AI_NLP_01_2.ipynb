```python
import nltk
nltk.download('wordnet')
```

```python
#copy + paste text / a paragraph into a string
text = """### Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language, in particular how to program computers to process and analyze large amounts of natural language data. The goal is a computer capable of "understanding" the contents of documents, including the contextual nuances of the language within them. The technology can then accurately extract information and insights contained in the documents as well as categorize and organize the documents themselves."""
text
```

```python
# with the slit() function we may create tokens
txt = text.split() # identifies words but not punctuation
txt
```

Tokenize

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')

tokens = word_tokenize(text.lower())
tokens
```

Remove punctuation

```python
import string

# create extended punctuation set to get rid of unwanted symbols
extended_punctuation = set(string.punctuation).union({'``', "''"})

# build a list comprehension with all tokens except for those in
# string.punctuation
tokens_without_punctuation = [w for w in tokens if w not in extended_punctuation]
tokens_without_punctuation
```

Remove stop words

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# create a list comprehension with all tokens_without_punctuation without
# the stopwords
tokens_without_stopwords = [w for w in tokens_without_punctuation if w not in stopwords.words()]
tokens_without_stopwords
```

Stemming

```python
from nltk.stem.porter import PorterStemmer

stemmed_tokens = [PorterStemmer().stem(w) for w in tokens_without_stopwords]

stemmed_tokens
```

Lemmatizing

```python
from nltk.stem import WordNetLemmatizer
```

```python
tokens_lemmatized = [WordNetLemmatizer().lemmatize(w, "n") for w in tokens_without_stopwords ]

tokens_lemmatized
```

Frequency of words


```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# count the number of occurences of each term
counts = Counter(tokens_lemmatized).most_common(10)
counts
```

```python
# barplot
labels, values = zip(*counts)
```

```python
fig, ax = plt.subplots()
ax.bar(labels, values)
plt.xticks(rotation =45)
```

###Similarities

```python
# import packages for cosine similarity
import nltk
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
```

```python

# Tokenize
tokens1 = nltk.word_tokenize(content1.lower())
tokens2 = nltk.word_tokenize(content2.lower())

# Remove punctuation
# create extended punctuation set to get rid of unwanted symbols
import string
extended_punctuation = set(string.punctuation).union({'``', "''"})

tokens1_without_punctuation = [word for word in tokens1 if word not in extended_punctuation]
tokens2_without_punctuation = [word for word in tokens2 if word not in extended_punctuation]

```

Identify similarities

```python
# list of documents to check for similarity
documents = [tokens1_without_punctuation, tokens2_without_punctuation]

# model for calculation
model = Word2Vec(documents, min_count=1)

# create vectors
doc1_vector = model.wv[tokens1_without_punctuation]
doc2_vector = model.wv[tokens2_without_punctuation]

#identify similarities
similarity_scores = cosine_similarity(doc1_vector, doc2_vector)

# overall similarity
import numpy as np
overall_similarity = np.mean(similarity_scores)
similarity_scores
```

Sentiment analysis

```python
from textblob import TextBlob
```

```python
TextBlob("happy").polarity
```

```python
TextBlob("angry").polarity
```

```python
TextBlob("angry").subjectivity
```

```python
# Function to analyze the sentiment
def analyze_sentiment(txt):
  polarity = TextBlob(txt).polarity

  if polarity > 0:
    sentiment_label = "positive"

  elif polarity < 0:
    sentiment_label = "negative"

  else:
    sentiment_label = "neutral"

  return sentiment_label


```

```python
analyze_sentiment("school")
```

```python
for text in tokens1_without_punctuation:
  sentiment = analyze_sentiment(text)
  print(f"Text: '{text}' | Sentiment : '{sentiment}' ")
```
